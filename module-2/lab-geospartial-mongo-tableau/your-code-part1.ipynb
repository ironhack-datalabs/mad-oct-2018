1. From all the companies get first office, number of people and company name

```python
from pymongo import MongoClient 
client = MongoClient("mongodb://admin:admin@localhost:32770")
db = client.mybbdd
```

```python
first_office = db.docs.find({"$and": [{"number_of_employees":{"$ne":None}}, {"offices.latitude":{"$ne":None}}, {"offices.longitude": {"$ne":None}}, {"offices": {"$ne":[]}}]}, projection={"name": 1 , "number_of_employees": 1, '_id':0, "offices.latitude": 1, "offices.longitude": 1})
first_office = list(first_office)

a = []
for item in first_office:
    a.append([item['offices'][0]['latitude'], item['offices'][0]['longitude'], item['number_of_employees'], item['name']])
print(type(a))
print(a)
```

2. Convert it to GeoJSON format
3. Insert it back to another mongodb collection

```python
def get(geojson):
    for i in geojson:
        point = Point((i[0], i[1]))
        features = []
        features.append(Feature(geometry=point, properties={"number_of_employees": i[2], "name": i[3]}))
        feature_collection = FeatureCollection(features)
    #with open('myfile.geojson', 'w') as f:
        #dump(feature_collection, f)
    return features
get(a)
```

```python
def first_office_point(ls): 
    points = []
    for i in ls:
        point = Point((i[1], i[0]))
        points.append(point)
    with open('points.geojson', 'w') as f:
        dump(points, f)
    return points
first_office_point(a)
```

```python
def first_office_points(ls): 
    features = []
    for i in ls:
        point = Point((i[1], i[0]))
        features.append(Feature(geometry=point, properties={"number_of_employees": i[2], "name": i[3]}))
        feature_collection = FeatureCollection(features)
    #with open('myfile.geojson', 'w') as f:
        #dump(feature_collection, f)
    return feature_collection
first_office_points(a)
```
