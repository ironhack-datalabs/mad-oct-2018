```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bitcoin = pd.read_csv('bitcoin.csv')
display(bitcoin.head(3))
ethereum = pd.read_csv('ethereum.csv')
display(ethereum.head(3))
litecoin = pd.read_csv('litecoin.csv')
display(litecoin.head(3))
```

```python
infob=bitcoin.describe(include='all').T
infob['type']=bitcoin.dtypes
display(infob)
infoe=ethereum.describe(include='all').T
infoe['type']=ethereum.dtypes
display(infoe)
infol=litecoin.describe(include='all').T
infol['type']=litecoin.dtypes
display(infol)
```

```python
biteth = pd.merge(bitcoin, ethereum, on='Date')
bietli = pd.merge(biteth, litecoin, on='Date')
new_columns = {'Open_x':'OpenB',
               'High_x':'HighB',
               'Low_x':'LowB',
               'Close_x':'CloseB',
               'Volume_x':'VolumeB',
               'Market Cap_x':'Market CapB',
               'Open_y':'OpenE',
               'High_y':'HighE',
               'Low_y':'LowE',
               'Close_y':'CloseE',
               'Volume_y':'VolumeE',
               'Market Cap_y':'Market CapE',
               'Open':'OpenL',
               'High':'HighL',
               'Low':'LowL',
               'Close':'CloseL',
               'Volume':'VolumeL',
               'Market Cap':'Market CapL'}
bietli.rename(columns=new_columns,inplace=True)
bietli['Date'] = pd.to_datetime(bietli['Date'])
display(bietli.head(3))
```

```python
bietli['difB'] = [bietli['OpenB'][i]-bietli['CloseB'][i] for i in range(len(bietli['OpenB']))]
bietli.loc[bietli['difB'] > 0, 'difB'] = 1
bietli.loc[bietli['difB'] < 0, 'difB'] = -1
bietli['difE'] = [bietli['OpenE'][i]-bietli['CloseE'][i] for i in range(len(bietli['OpenE']))]
bietli.loc[bietli['difE'] > 0, 'difE'] = 1
bietli.loc[bietli['difE'] < 0, 'difE'] = -1
bietli['difL'] = [bietli['OpenL'][i]-bietli['CloseL'][i] for i in range(len(bietli['OpenL']))]
bietli.loc[bietli['difL'] > 0, 'difL'] = 1
bietli.loc[bietli['difL'] < 0, 'difL'] = -1
bietli.head(3)
```
