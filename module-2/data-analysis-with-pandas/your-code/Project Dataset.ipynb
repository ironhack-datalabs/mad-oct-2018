### Restaurants & Cuisine 
El segundo dataset que queremos limpiar y modificar para obetener los datos que queremos es los tipos de comida de cada restaurante. 
Para ello debemos de limpiar las categorías y renombrar categorías más globales, una por cada restaurante. 

````python
# Exportamos dataset
restaurant_type = pd.read_csv('chefmozcuisine.csv')
display(restaurant_type.head())

````

````python
restaurant_type.groupby('placeID')
````

````python
restaurant_type.info()
````

````python
# Modificamos el tipo de objeto de Rcuisine a Categórico. 
restaurant_type[restaurant_type.select_dtypes(['object']).columns] = restaurant_type.select_dtypes(['object']).apply(lambda x: x.astype('category'))
````

````python
#Vemos todas la categorías.
restaurant_type['Rcuisine'].cat.categories
````

````python
#Definimos una función para concatenar los valores de Rcuisine
def Rcousine (df):
    return (' , '.join(df['Rcuisine']))
````

````python
# Aplicamos la función a nuestro dataframe y renombramos la columnas
restaurant_type_agg = pd.DataFrame(restaurant_type.groupby('placeID', as_index=True).apply(Rcousine))
restaurant_type_agg.columns = ['Food_type']
restaurant_type_agg.head()
````

````python
#Podemos unir el dataset de restaurantes con el dataset de tipos de comida
rest_new = pd.merge(restaurants, restaurant_type_agg, on='placeID')
rest_new.info()
````

Comprobamos que rest_new, nuestro nuevo dataframe donde tenemos toda la información del dataset de "restaurants" más toda la info del tipo de comida, solo nos devuelve 76 entradas. Podemos concluir que solo 76 de los 130 restaurantes que tenemos tiene información sobre el tipo de comida.
