```python
import pandas as pd
import pymongo
from pandas.io.json import json_normalize
```

```python
from pymongo import MongoClient 
client = MongoClient("mongodb://admin:admin@localhost:32768")
db = client.mybbdd
```

```python
docs = db.docs.find({"founded_year":{"$gte":2010}})
print(len(list(docs)))
```

```python
# 1. All the companies that it's name match 'Babelgum'. Retrieve only their name field.
companies1 = db.docs.find({"name":{"$eq":"Babelgum"}}, { "name": 1 , "_id": 0})
companies1[0]
```

```python
# 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by number of employees.
companies2 = db.docs.find({"number_of_employees":{"$gte":500}}, limit=20, projection={ "name": 1 , "number_of_employees": 1, "_id": 0}).sort("number_of_employees", pymongo.DESCENDING)
companies2 = pd.DataFrame(list(companies2))
companies2
```

```python
# 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the name and founded_yearfileds.
companies3 = db.docs.find({'founded_year':{'$gte':2000, '$lte':2005}}, projection={'name':1, 'founded_year':1, '_id':0})
companies3 = pd.DataFrame(list(companies3))
companies3
```

```python
#4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the name and ipo fields.
companies4 = db.docs.find({'$and': [{'ipo.valuation_amount': {'$gt':100000000}}, {'founded_year': {'$lt':2010}}]}, projection={'name':1, 'ipo':1, '_id':0})
companies4 = json_normalize(list(companies4))
companies4
```

```python
# 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.
companies5 = db.docs.find({'$and':[{"number_of_employees":{"$lt":1000}},{'founded_year':{'$lt':2005}}]}, limit=10, projection={'name':1, 'founded_year':1, 'number_of_employees':1}).sort('number_of_employees', pymongo.DESCENDING)
companies5 = pd.DataFrame(list(companies5))
companies5
```

```python
# 6. All the companies that don't include the partners field.
companies6 = db.docs.find({'partners':{'$eq':[]}}, limit=200, projection={'name':1, 'partners':1})
companies6 = pd.DataFrame(list(companies6))
companies6
```

```python
#7. All the companies that have a null type of value on the category_code field.
companies7 = db.docs.find({'category_code':{'$type':['null']}}, limit=200, projection={'name':1, 'category_code':1, '_id':0})
companies7 = pd.DataFrame(list(companies7))
companies7
```

```python
# 8. All the companies that have at least 100 employees but less than 1000. Retrieve only the name and number of employees fields.
companies8 = db.docs.find({'number_of_employees':{'$gte': 100, '$lt': 1000}}, projection={'name':1, 'number_of_employees':1})
companies8 = pd.DataFrame(list(companies8))
companies8
```

```python
# 9. Order all the companies by their IPO price descendently.
companies9 = db.docs.find({'ipo.valuation_amount':{'$type':'number'}}, projection={"name":1 ,"ipo.valuation_amount": 1}).sort("ipo.valuation_amount", pymongo.DESCENDING)
companies9 = json_normalize(list(companies9))
companies9
```

```python
# 10. Retrieve the 10 companies with more employees, order by the number of employees
companies10 = db.docs.find({'number_of_employees':{'$gte': 0}}, limit=10, projection={'name':1, 'number_of_employees':1}).sort('number_of_employees', pymongo.DESCENDING)
companies10 = pd.DataFrame(list(companies10))
companies10
```

```python
# 11. All the companies founded on the second semester of the year. Limit your search to 1000 companies.
companies11 = db.docs.find({'founded_month':{'$gte': 6}}, limit=100, projection={'name':1, 'founded_month':1})
companies11 = json_normalize(list(companies11))
companies11
```

```python
# 13. All the companies founded before 2000 that have and acquisition amount of more than 10.000.000
companies13 = db.docs.find({'$and' : [{"founded_year":{'$lt':2000}}, {"acquisition.price_amount":{'$gt':10000000}}]},projection={"name": 1 , "founded_year": 1, "acquisition.price_amount": 1})
companies13 = json_normalize(list(companies13))
companies13
```

```python
# 14. All the companies that have been acquired after 2015, order by the acquisition amount, and retrieve only their nameand acquisiton field.
companies14 = db.docs.find({[{"acquisition.acquired_year":{'$lt':2015}}]},projection={"name": 1, "acquisition.acquired_year": 1, '_id':0}).sort("acquisition.price_amount", pymongo.DESCENDING)
#companies14 = pd.DataFrame(list(companies14))
companies14
```
