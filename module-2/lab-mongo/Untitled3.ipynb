```python
from pymongo import MongoClient
import pandas as pd
from pandas.io.json import json_normalize
```

```python
client=MongoClient("mongodb://admin:admin@localhost:32772")
db=client.ccc
```

1. All the companies that it's name match 'Babelgum'. Retrieve only their name field.

```python
query1=db.companies.find({"name":{"$eq":"Babelgum"}},projection={"name":True, "_id":False})
```

```python
df1= pd.DataFrame(list(query1))
df1
```

2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by number of employees.

```python
query2=db.companies.find({"number_of_employees":{"$gt":5000}},projection={"name":1, "number_of_employees":1, "_id":0}).sort("number_of_employees", -1).limit(20)
df2= pd.DataFrame(list(query2))
df2
```
