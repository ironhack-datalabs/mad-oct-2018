```python
# https://docs.mongodb.com/manual/reference/operator/query/
from pymongo import MongoClient
import pymongo
import pandas as pd
from pandas.io.json import json_normalize
```

```python
client = MongoClient("mongodb://admin:admin@localhost:32768")
db = client.mybbdd
```

```python
# docs = db.docs.find({"founded_year":{"$gte":2010}})
# print(len(list(docs)))
```

1. All the companies that it's name match 'Babelgum'. Retrieve only their name field.

```python
# name
companies1 = db.docs.find({"name":{"$eq":"Babelgum"}}, { "name": 1 , "_id": 0})
df1 = pd.DataFrame(list(companies1))
df1.head()
```

2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by number of employees.

```python
# number_of_employees
companies2 = db.docs.find({"number_of_employees":{"$gt":5000}}, 
                          limit=20, 
                          projection={ "name": 1 , "number_of_employees": 1, "_id": 0}).sort('number_of_employees' , pymongo.DESCENDING)
df2 = pd.DataFrame(list(companies2))
print(len(df2))
df2
```

3. All the companies founded between 2000 and 2005, both years included. Retrieve only the name and founded_yearfileds.

```python
# founded_year
companies3 = db.docs.find({"founded_year":{"$gte":2000, "$lte":2005}}, 
                          projection={"name": 1 , "founded_year": 1, "_id": 0})

d = list(companies3)
df3 = pd.DataFrame(d)
print(len(df3))
df3.head(10)
```

4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the name and ipo fields.

```python
# Valuation Amount
# hay que buscar ipo en compass
companies4 = db.docs.find({"founded_year":{"$lt":2010}, "ipo.valuation_amount": {"$gte": 100000000}}, 
                          projection={"name": 1 , "ipo.valuation_amount": 1, "_id": 0})


d = list(companies4)

# pd4 = json_normalize((list(doc4)))

df4 = pd.DataFrame(d)
# print(len(df4))
df4['ipo'] = json_normalize(df4['ipo'])

df4.head(10)
```

5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
# number_of_employees
companies5 = db.docs.find({"number_of_employees":{"$lt":1000}, "founded_year": {"$lt": 2005}}, 
                          limit=10, 
                          projection={"name": 1 , "number_of_employees": 1, "_id": 0}).sort('number_of_employees' , pymongo.DESCENDING)

d = list(companies5)
df5 = pd.DataFrame(d)
df5
```

6. All the companies that don't include the partners field.

```python
# not partners field
companies6 = db.docs.find({"partners":{"$exists":False}}, 
                          # limit=10, 
                          projection={"name": 1, "_id": 0})

d = companies6
# df6 = pd.DataFrame(d)
# df6
print(len(list(d)))

# Todos tienen campo partners
```

7. All the companies that have a null type of value on the category_code field.

```python
# not category_code field
companies7 = db.docs.find({"category_code":{"$eq":None}}, 
                          # limit=10, 
                          projection={"name": 1, "_id": 0})

d = companies7
# df7 = pd.DataFrame(d)
# df7
print(len(list(d)))
```

8. All the companies that have at least 100 employees but less than 1000. Retrieve only the name and number of employees fields.

```python
# not category_code field
companies8 = db.docs.find({"number_of_employees":{"$gt":100, "$lt": 1000}}, 
                          # limit=10, 
                          projection={"name": 1, "number_of_employees": 1, "_id": 0})

# df7 = pd.DataFrame(d)
# df7
d = list(companies8)
print(len(d))
df8 = pd.DataFrame(d)
df8.head()
```

9. Order all the companies by their IPO price descendently.

```python
### no sale

'''
doc9 = coll.find({"ipo.valuation_amount":{"$ne":None}},projection={"name":1,"ipo.valuation_amount":1,"_id":0}).sort("ipo.valuation_amount",pymongo.DESCENDING)
'''


companies9 = db.docs.find({"ipo.valuation_amount":{"$ne":None}},
                          projection={"name":1,"ipo.valuation_amount":1,"_id":0}).sort("ipo.valuation_amount",pymongo.DESCENDING)

# df7 = pd.DataFrame(d)
# df7
d = list(companies9)
print(len(d))
df9 = pd.DataFrame(d)
df9['ipo'] = json_normalize(df9['ipo'])

df9.head()
```

10. Retrieve the 10 companies with more employees, order by the number of employees

```python
# number_of_employees
companies10 = db.docs.find( 
                          limit=10, 
                          projection={"name": 1, "number_of_employees": 1, "_id": 0}).sort('number_of_employees' , pymongo.DESCENDING)

d = list(companies10)
df10 = pd.DataFrame(d)
df10
```

11. All the companies founded on the second semester of the year. Limit your search to 1000 companies.

```python
# founded_month
companies11 = db.docs.find( {"founded_month": {"$in": [7, 8, 9, 10, 11, 12]}}, 
                          limit=1000, 
                          projection={"name": 1, "_id": 0})

d = list(companies11)
df11 = pd.DataFrame(d)
df11.head(20)
```

12. All the companies that have been 'deadpooled' after the third year.

```python
# deadpooled

# no sale 


companies12 = db.docs.find( # { $subtract: [ "deadpooled_year", "founded_year" ] }, 
                            {"deadpooled_year":{"$gte":3, "$lt": 1000}}, 
                          limit=10, 
                          projection={"name": 1, "deadpooled_year": 1, "_id": 0})

d = list(companies12)
df12 = pd.DataFrame(d)
df12.head(20)
```

13. All the companies founded before 2000 that have and acquisition amount of more than 10.000.000

```python
# founded 

# {"founded_year":{"$lt":2010}, "ipo.valuation_amount": {"$gte": 100000000}}

companies13 = db.docs.find( {"founded_year":{"$lt":2000},  
                             "acquisition.price_amount": {"$gt": 10000000}}, 
                           projection={"name": 1, "founded_year": 1, "acquisition.price_amount": 1, "_id": 0})

d = list(companies13)

# json_normalize(d['programs'])

df13 = pd.DataFrame(d)
df13['acquisition'] = json_normalize(df13['acquisition'])
df13
```

14. All the companies that have been acquired after 2015, order by the acquisition amount, and retrieve only their nameand acquisiton field.

```python
# adquired 2015 

# Cambiamos los datos de 2015 -> 2010
companies14 = db.docs.find( {"acquisition.acquired_year": {"$gt": 2010}}, 
                           projection={"name": 1, "acquisition.price_amount": 1, "_id": 0}).sort("acquisition.price_amount", pymongo.DESCENDING)

d = list(companies14)
print(len(d))

df14 = pd.DataFrame(d)
df14['acquisition'] = json_normalize(df14['acquisition'])
df14.head()
```

15. Order the companies by their founded year, retrieving only their name and founded year.

```python
# founded year

companies15 = db.docs.find(limit=1000, 
                           projection={"name": 1, "founded_year": 1, "_id": 0}).sort("founded_year", pymongo.DESCENDING)

d = list(companies15)
print(len(d))

df15 = pd.DataFrame(d)
# df15['acquisition'] = json_normalize(df15['acquisition'])
df15.head(15)
```

16. All the companies that have been founded on the first seven days of the month, including the seventh. Sort them by their aquisition price descendently. Limit the search to 10 documents.

```python
# founded year

companies16 = db.docs.find({"founded_day": {"$in": [1,2,3,4,5,6,7]}}, 
                           limit=10, 
                           projection={"name": 1,"_id": 0}).sort("acquisition.price_amount", pymongo.DESCENDING)

d = list(companies16)
print(len(d))

df16 = pd.DataFrame(d)
# df15['acquisition'] = json_normalize(df15['acquisition'])
df16
```

17. All the companies on the 'web' category that have more than 4000 employees. Sort them by the amount of employees in ascendant order.

```python
# web

companies17 = db.docs.find({"number_of_employees":{"$gt":4000}, 
                            "category_code": {"$eq": "web"}}, 
                           projection={"name": 1, "number_of_employees": 1, "_id": 0}).sort("number_of_employees", pymongo.DESCENDING)

d = list(companies17)
print(len(d))

df17 = pd.DataFrame(d)
# df15['acquisition'] = json_normalize(df15['acquisition'])
df17
```

18. All the companies which their acquisition amount is more than 10.000.000, and currency are 'EUR'.

```python
# EUR

companies18 = db.docs.find({"acquisition.price_amount":{"$gt":10000000}, 
                            "acquisition.price_currency_code": {"$eq": "EUR"}}, 
                           projection={"name": 1, "_id": 0})

d = list(companies18)
print(len(d))

df18 = pd.DataFrame(d)
# df15['acquisition'] = json_normalize(df15['acquisition'])
df18
```

19. All the companies that have been acquired on the first trimester of the year. Limit the search to 10 companies, and retrieve only their name and acquisition fields.

```python
# trimester

companies19 = db.docs.find({"acquisition.acquired_month":{"$in":[1,2,3]}}, 
                           limit = 10, 
                           projection={"name": 1, "acquisition": 1, "_id": 0})

d = list(companies19)
print(len(d))

df19 = pd.DataFrame(d)
# df19['acquisition'] = json_normalize(df19['acquisition'])
df19
```

20. All the companies that have been founded between 2000 and 2010, but have not been acquired before 2011.

```python
# founded adquired

# field: { $not: { <operator-expression> } }

companies20 = db.docs.find({"founded_year": {"$gte": 2000, "$lte": 2010}, 
                            "acquisition.acquired_year":{"$not":{"$lt": 2011} }}, 
                           projection={"name": 1, "founded_year": 1, "acquisition.acquired_year": 1, "_id": 0})

d = list(companies20)
print(len(d))

df20 = pd.DataFrame(d)
# df19['acquisition'] = json_normalize(df19['acquisition'])
df20
```
