#### Challenge 3: Hidden Cold Joke

Using Python, call Github API to find out the cold joke contained in the 24 secret files in this repo whose filenames contain .scavengerhunt. The secret files are named from .0001.scavengerhunt to .0024.scavengerhunt. They are scattered randomly throughout this repo. You need to search for these files by calling the Github API, not searching on your computer.

```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
```

```python
api='https://api.github.com/repos/'
user='ironhack-datalabs/madrid-oct-2018/'
what='git/trees/master'
how='?recursive=1'
completerequest=api+user+what+how
response = requests.get(completerequest)
result = response.json()
display(result)
```

```python
scanvengers=[]
scv=[]
for item in pd.DataFrame(result['tree'])['path']:
    if str(item).endswith('scavengerhunt'):
        scanvengers.append('../madrid-oct-2018/'+item)
        if '/' in item:
            scv.append(item.split('/')[-1])
        else:
            scv.append(item)
```

```python
final_df=pd.DataFrame(scanvengers,columns=['path'])
final_df['names']=scv
final_df=final_df.sort_values('names',ascending=True)
display(final_df)
```
