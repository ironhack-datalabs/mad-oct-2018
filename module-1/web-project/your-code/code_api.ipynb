- I consulted the twitter API for tweets containing "friyay" using the search method

```python
friyay = api.search("friyay")
```

```python
tweets = pd.DataFrame([pd.Series(tweet._json) for tweet in friyay])
print(type(tweets))
tweets.head()
```
