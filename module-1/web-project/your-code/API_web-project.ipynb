````python
#Importamos las librerías principales para esta práctica
import json
import requests
import pandas as pd
import numpy as np
from pandas.io.json import json_normalize
````

### Importamos datos

````python
#A través de response, importamos todos los datos. 
response = requests.get('http://makeup-api.herokuapp.com/api/v1/products.json')
````

````python
#Normalizamos nuestros datos
results = response.json()
df = json_normalize(results)
df.head()
````

````python
#Eliminamos columnas innecesarias 
drop_cols = ['api_featured_image', 'product_api_url', 'image_link']
df = df.drop(drop_cols, axis = 1)
df.head()
````

````python
#Ordenamos nuestras columnas. 
col_order = ['id', 'name', 'brand','product_type', 'price', 'price_sign', 'rating', 'description', 'tag_list', 'website_link', 'product_link', 'updated_at', 'product_colors']
df = df[col_order]
df.head()
````

### Definición de función

````python
#Extraemos los valores únicos de la columna "brand" y creamos una lista 
brand_list = df.brand.unique().tolist()
````

````python
#Extraemos los valores únicos de la columna "product_type" y creamos una lista 
product_type_list= df.product_type.unique().tolist()
````

````python
#Definimos una función para obtener la marca y los tipos de producto por marca. 
def consult (data):
    print("Brands: ", (brand_list))
    print("Product type: ", (product_type_list))
    brand_q = input("Please enter a brand: ")
    product_q = str(input("Please enter a product type: "))
    if brand_q.lower() in brand_list:
        if product_q.lower() in product_type_list:
            return(data.loc[(df['brand'] == brand_q) & (data['product_type'] == product_q)])
        else:
            return ("The brand or type of product are not available on the website")
````

````python
#Llamamos a la función con los datos del dataframe de la API
consult(df)
````
