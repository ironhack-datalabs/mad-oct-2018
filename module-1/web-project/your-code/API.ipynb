```python
import json
import requests
import pandas as pd

```

```python
response = requests.get('http://makeup-api.herokuapp.com/api/v1/products.json?brand')
results = response.json()
```

```python
#response = requests.get('http://makeup-api.herokuapp.com/api/v1/products.json?brand=' + brand)
```

```python
from pandas.io.json import json_normalize

results = response.json()
df = json_normalize(results)
df.info()
df
```

```python
#Extraigo tags a columnas
df_tag= pd.DataFrame(df.tag_list.values.tolist(), df.index).add_prefix('tag_')
print(df_tag)
```

```python
#Concateno DataFrames
df = pd.concat([df,df_tag],axis=1)
df.info()
display(df.head())
```

```python
# Cambio tipo de datos
df['category'] = df['category'].astype('category')
df['price'] = df['price'].astype('float')
df['tag_0'] = df['tag_0'].astype('category') 
df['tag_1'] = df['tag_1'].astype('category')
df['tag_2'] = df['tag_1'].astype('category')
df['tag_3'] = df['tag_1'].astype('category')
df['tag_4'] = df['tag_1'].astype('category')
df['tag_5'] = df['tag_1'].astype('category')
df['tag_6'] = df['tag_1'].astype('category')
df['tag_7'] = df['tag_1'].astype('category')
```

```python
df.info()
```

```python
### aqui veo todos los elementos de cada uno de nuestros tag_list

print(df['tag_0'].cat.categories)
print(df['tag_1'].cat.categories)
print(df['tag_2'].cat.categories)
print(df['tag_2'].cat.categories)
print(df['tag_4'].cat.categories)
print(df['tag_5'].cat.categories)
print(df['tag_6'].cat.categories)
print(df['tag_7'].cat.categories)
```

```python
## sacar el precio medio segun la catagoria y teniendo que estar en la misma divisa 
```

```python
avg_mpg = df.groupby(['category', 'currency'], as_index=False).agg({'price':'mean'})

display(avg_mpg)
```

```python
avg_mpg = df.groupby(['category'], as_index=False).agg({'rating':'mean'})

display(avg_mpg)
```
