```python
# Necesitamos exportar 2 librerias
import pandas as pd
import numpy as np
```

```python
# Importamos nuestra database desde un fichero local .csv.
data = pd.read_csv('BlackFriday/BlackFriday.csv')
```

```python
# Comprobamos los primeros primeros registros para conocer nuestra base de datos
display(data.head())
```

```python
# Con esta funci√≥n comprobamos los registros nulos y datatype de cada una de las columnas. 
data.info()
```

```python
# Queremos eliminar las columnas 'Stay_In_Current_City_Years', 'Product_Category_1' 'Product_Category_2' y 'Product_Category_3'

blackfriday = data[['User_ID', 'Product_ID', 'Gender', 'Age', 'Occupation', 'City_Category', 'Marital_Status', 'Purchase']]
display(blackfriday.head())
```

```python
#Renombramos columnas. 
blackfriday = blackfriday.rename(columns={'Purchase':'Expenditure',
                              'City_Category': 'City'})
display(blackfriday.head())
```

```python
#Comprobar tipo de dato
blackfriday.info()
```

```python
#Renombramos datos en 'City' creando una nueva columna
blackfriday.loc[blackfriday['City'].str.startswith('A'), 'City_name'] = 'Madrid'
blackfriday.loc[blackfriday['City'].str.startswith('B'), 'City_name'] = 'Barcelona'
blackfriday.loc[blackfriday['City'].str.startswith('C'), 'City_name'] = 'Sevilla'
display(blackfriday.head())
```

```python
# Para poder asignar "Single" y "Married" a los valores 0 y 1 de la columna "Marital_Status" tenemos que cambiar el tipo de datos. 
blackfriday['Marital_Status'] = blackfriday['Marital_Status'].astype('str')
blackfriday.loc[blackfriday['Marital_Status'].str.startswith('0'), 'Status'] = 'Single'
blackfriday.loc[blackfriday['Marital_Status'].str.startswith('1'), 'Status'] = 'Married'
display(blackfriday.head())
```

```python
#Eliminamos las columnas de "Marital_Status" y "City"
blackfriday = blackfriday.drop(blackfriday.columns[[5,6]], axis=1)
display(blackfriday.head())
```

```python
#Reordenamos las columnas
column_order = ['User_ID', 'Product_ID', 'Gender', 'Age', 'City_name', 'Occupation', 'Status', 'Expenditure']
blackfriday = blackfriday[column_order]
display(blackfriday.head())
```

```python
# Comprobamos si tenemos valores nulos. 
null_cols = blackfriday.isnull().sum()
print(null_cols)
```

```python
#Para poder conocer la distribucion de nuestros datos, calculamos mean, min, max, Q1 y Q3. Sabemos 
blackfriday['Expenditure'].max()
blackfriday['Expenditure'].min()
blackfriday['Expenditure'].mean()
Q1 = blackfriday['Expenditure'].quantile(0.25)
Q3 = blackfriday['Expenditure'].quantile(0.75)

#Sabemos que el 50% de los valores del gasto en compra se encuentran entre 5866 y 12073 (referencia en readme.md de foto).
```

```python
# Definimos los niveles de corte
expend_labels = ['Low', 'Moderate', 'High', 'Very High']
```

```python
# DDefinimos los valores de corte de manera uniforme
cutoffs = [0,6000, 12000, 18000, 24000]
bins = pd.cut(blackfriday['Expenditure'],cutoffs, labels=expend_labels)
print(bins.head())
```

```python
bins.value_counts()
#Aproximadamente 50% de los usuarios han tenido un nivel de gasto moderado. 

```

```python
black_f = pd.DataFrame(blackfriday)
black_f['Expend_type']=bins
display(black_f.head())
```

```python
#Perfil de consumidor mas concurrido
black_f.groupby(['Expenditure','Age', 'Status', 'Expend_type']).size().reset_index().max()
```

```python
# Single vs. Married
black_f.groupby(['Expend_type','Status']).size().reset_index()
```

```python
# Expend_type por Edad
black_f.groupby(['Age','Expend_type']).size().reset_index()
```

```python
# Expend_type por Edad
black_f.groupby(['Expend_type','Age']).size().reset_index()
```
