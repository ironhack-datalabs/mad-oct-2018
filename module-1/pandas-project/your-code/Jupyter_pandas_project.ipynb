```python
import pandas as pd
import numpy as np
```

```python
##### Import the DATA 2016 
data16 = pd.read_csv('world-happiness-report/2016.csv')
data16
```

```python
####Import the DATA table 2017
data17 = pd.read_csv('world-happiness-report/worldhappines.csv')
data17
```

```python
## AÃ±ado una columna en ambas datas que me indique el aÃ±o de los datos de cada fila para que al hacer el concat pueda diferenciarlos

data17['year'] = '2017'
data17
data16['year'] = '2016'
data16
```

```python
#### He cambiado los nombres de la table de 2017 para que sean exactamente iguales que los de la data de 2016 y coincidan.

data17 = data17.rename(columns={'Happiness.Rank': 'Happiness Rank','Happiness.Score': 'Happiness Score', 'Economy..GDP.per.Capita.' :'Economy (GDP per Capita)','Trust..Government.Corruption.': 'Trust (Government Corruption)'})
data17.head()
```

```python
#### Realizo unas datas nuevas con las columnas que me van a ser utiles par realizar la conclsion

data16 = data16[['Country','Happiness Rank','Happiness Score','Economy (GDP per Capita)','year']]
data17 = data17[['Country','Happiness Rank','Happiness Score','Economy (GDP per Capita)','year']]
display(data16.head())
display(data17.head())
```

```python
#### Concateno ambas tablas con las columnas que me interesan, y creo unas keys para indicar donde empieza el nuevo aÃ±o.

df=pd.concat([data16, data17 ], keys=['2016', '2017',])
df

```

```python
#### Bins
mpg_labels = ['Low','Medium','High']
bins = pd.cut(df['Economy (GDP per Capita)'],3, labels=mpg_labels)
bins.head(100)
```

```python
bins.value_counts()
```

```python
### aÃ±adir los bins como columnas de la base de datos 

df['BINS'] = bins
df
```

```python
###### Conclusion: Con este analisis podemos concluir que el nivel de felicidad para algunos paises depende de algun modo al nivel de felicidad del pais. Es decir, el ranking esta liderado por paises que se situan en una posicion high en funcion a la aportacion de GDP a la felicidad de las personas. pero por otro lado tambien podemos encontrar casos exclusivos como Costa Rica y Brazil, los cuales a pesar de que su aportacion de gdp al nivel de Happines es menos que muchos paises se encuentran en un posicion bastante alta lo que indica que para ellos el GDP no es un factor determinante para su felicidad
##### Tambin podemos obsevar como los paises menos felices (los ultimos del ranking) la grana mayoria cuentan con una aportacion low por parte del GDP por lo que de algun modo podemos definirlo como determinante para ello.
```

```python
### He creado dos tablas a partir de las colunnas de 'Country' y 'Happiness Ranking' para poder comparar el aumento o decrease en la posicion de la ciudad en el Happiness ranking conforme al aÃ±o anterior 
rank16 = data16[['Country','Happiness Rank']]
rank16
rank17 = data17[['Country','Happiness Rank']]
rank17
```

```python
### Esta comparacion la he hecho a traves de un merge dosnde me ha posicionado cada 'country' segun el aÃ±o en la posicion 

merge = pd.merge(rank16, rank17, on=['Country'], how='outer')
merge.head()
```
