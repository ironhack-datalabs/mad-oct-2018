## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
import numpy as np
import os
```

```python
# Listing files
rootfold=os.getcwd().split('/your-code')[0]
g=[rootfold+"/"+f for f in os.listdir(rootfold) if '.csv' in f]
filename=np.unique(g)[0]
print(filename)
```

```python
weather=pd.read_csv(filename)
```

```python
weather
```

```python
# Number of NULL values for weather Data Frame per Column
display(weather.isnull().sum())
```

We can observe that the only columns that have NULL values are those with the identifier d before its nuber assigment<br>

```python
dlist=[] # Names of the columns with relevant data of temperature
# We are going to work with columns without data 'd#'
for item in weather.columns.tolist():
    if item.startswith('d'):
        dlist.append(item)
```

```python
tempdf=weather[dlist]
tempdf.describe().T
```

The analysis of d# columns throws us the idea that there are many sensors that are not working properly, <br>
count equal to 0, and the receptors with more than 0 counts are far away from the number of total registers<br>
of the data frame.<br>
Next **step** consists on applying a mean horizontaly to our values, to obtain an overall value of the temperature<br>
that allow us to gather some useful data.

```python
meantemp=tempdf.mean(numeric_only=True,axis=1)
display(meantemp)
```

We can add now the values to the original data frame, creating a secondary data frame called **weather_2**<br>
Then we will divide data for tmin and tmax for column **element**

```python
weather_2=weather.copy()
weather_2['Mean Temp']=meantemp
display(weather_2)
```

### Using matplotli.pyplot to plot min and max temperatures

```python
# Uncomment the next line to install matplotlib.pyplot
# !pip3 install matplotlib
import matplotlib.pyplot as plt
```

```python
# Min temp array
min_temp=weather_2[weather_2['element']=='tmin']
min_temp=np.array(min_temp['Mean Temp'])
or_min_temp=weather[weather['element']=='tmin'][dlist].reset_index(drop=True).as_matrix()
# Max temp array
max_temp=weather_2[weather_2['element']=='tmax']
max_temp=np.array(max_temp['Mean Temp'])
or_max_temp=weather[weather['element']=='tmax'][dlist].reset_index(drop=True).as_matrix()
```

```python
plt.close('all')

plt.figure(figsize=(20,10))
x=weather['month'].unique()

plt.plot(x,max_temp,color='red',label='Mean Max Temp')
plt.plot(x,min_temp,color='blue',label='Mean Min Temp')
for ind in range(np.size(or_max_temp,0)):
    temp=or_max_temp[ind,:]
    temp1=or_max_temp[ind,:]
    xt=np.zeros(len(temp))
    xt[:]=x[ind]
    if ind==np.size(or_max_temp)-1:
        plt.plot(xt,temp,'dr',label='scatter original Max temp')
        plt.plot(xt,temp1,'*b',label='scatter original Min temp')
    else:
        plt.plot(xt,temp,'*r')
        plt.plot(xt,temp1,'*b')  
plt.grid()
plt.legend()

plt.xlabel('Month')
plt.ylabel('Temperature (Celsius)')
plt.title('Historic of temperature for '+str(weather['id'].unique()[0])+' station',fontweight='bold')

plt.show()
```
